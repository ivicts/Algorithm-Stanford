```python
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

```python
def read_input(filename):
    first = True
    counter = 0 
    with open('week14_file/'+filename) as f:
        for line in f:
            line = line.split() # to deal with blank 
            if line and not first:            # lines (ie skip them)
                x = float(line[0])
                y = float(line[1])
                TSP_instance[counter,0] = x
                TSP_instance[counter,1] = y
                counter+=1
            else:
                num_cities = int(line[0])
                TSP_instance = np.zeros((num_cities,2))
                first = False
    
    #Compute the distance between two cities and store in hash table for faster lookup
    G = {}

    for i in range(num_cities):
        for j in range(num_cities):
            if i<j:
                x = TSP_instance[i][0]
                y = TSP_instance[i][1]
                z = TSP_instance[j][0]
                w = TSP_instance[j][1]
                dist = np.sqrt((x-z)**2+(y-w)**2)
                G[(i,j)] = dist
                
    return TSP_instance,num_cities,G
```

```python
def travelling_salesman_problem(filename=None,TSP_instance=None):
    """Running time O(n^2*2^n)"""
    #Either using list or read a txt file
    if filename is not None and TSP_instance is None:
        TSP_instance,num_cities,G = read_input(filename)
    elif filename is None and TSP_instance is not None:
        num_cities = len(TSP_instance)     
        #Compute the distance between two cities and store in hash table for faster lookup
        G = {}
        for i in range(num_cities):
            for j in range(num_cities):
                if i<j:
                    x = TSP_instance[i][0]
                    y = TSP_instance[i][1]
                    z = TSP_instance[j][0]
                    w = TSP_instance[j][1]
                    dist = np.sqrt((x-z)**2+(y-w)**2)
                    G[(i,j)] = dist
    
    #index goes from 0 to n-1, 0 is the source 
    #For each set S subsets of {0,1,...,n-1} of any size that contains 0
    S = [x for x in powerset(list(range(num_cities))) if 0 in x]
    
    #Base case
    #A is hash table with key value is a tuple (S,j)
    #where S is a subets of {0,1,...,n-1} that contain 0 and j is a destination element {0,1,...,n-1}
    A_old = {} 
    A_old[(0,),0] = 0 
    
    #m = subproblem size
    for m in range(2,num_cities+1):  

        A_new = {}
        #For each set S subsets of {0,1,...,n-1} of size m that contains 0
        subsets_S = [x for x in S if len(x) == m]
        for subset_S in subsets_S:
            A_new[subset_S,0] = np.inf #No way to avoid visiting vertex (twice)
            list_subset_S = list(subset_S)
            #for each j element S and j is not 0
            for j in subset_S:
                if j is not 0:
                    wkj = []
                    for k in subset_S:
                        if k is not j:
                            if k<j:
                                ckj = G[(k,j)]
                            else:
                                ckj = G[(j,k)]
                         
                            subset_S_j = [x for x in list_subset_S if x != j]
                            wkj.append(A_old[(tuple(subset_S_j),k)]+ckj)
                            
                    #Let P be a shortest path from 0 to j that visits the vertices S (assume |S| > = 2)
                    #[exactly one each]. If last hop of P is (k,j), then P' is a shortest path from 1 to k 
                    #that visits every vertex of S-{j} exactly once.
                    A_new[(subset_S,j)] = min(wkj)
                    
        #for space optimization, only save the dictionary value when it is needed
        A_old = A_new
    wj0 = []
    
    for j in range(1,num_cities):
        #min cost from 0 to j visiting everybody once + cost of final hop of tour
        wj0.append(A_new[tuple(list(range(num_cities))),j]+G[(0,j)])
    
    return min(wj0)  
```

```python
start_time = time.time()
print(travelling_salesman_problem('week14_test1.txt')) #10.24
print("--- %s seconds ---" % (time.time() - start_time))
```

We can find the real answer to the TSP problem - a minimum-cost tour (i.e, a cycle that visits every vertex exactly once), by calculating the TSP of the city 0-12 + TSP of the city 11-24 - 2*distance between city 11 and 12. Since in the graph we can see that we double count the line that connects city 11 and 12, and we do not actually need this line in the original TSP of the city 0-24 because it will create a cycle, so we need to substract twice of the distance between city 11 and 12.

```python
start_time = time.time()
print(travelling_salesman_problem(TSP_instance=TSP_instance[:13])+travelling_salesman_problem(TSP_instance=TSP_instance[11:])-2*G[(11,12)]) #26442.730309
print("--- %s seconds ---" % (time.time() - start_time))
```
