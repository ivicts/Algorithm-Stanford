```python
def bellman_ford(G_rev,num_vertices,num_edges,source):
    
    #num_vertices+1 to check cycle
    A = np.zeros((num_vertices+1,num_vertices))
    
    #Intialize base cases
    A[0,:] = np.inf
    A[0,source] = 0
    
    #iterate num_vertices+1 times to check cycle
    for i in range(1,num_vertices+1):
        for v in G_rev.keys():
            wv = []
            for edges in G_rev[v]:
                #print(v,edges)
                #print(edges)
                w = edges[0]
                cwv = edges[2]
                wv.append(A[i-1,w]+cwv)
            #print(wv)
            A[i,v] = min(A[i-1,v],min(wv))
    
    #print(A)
    if np.array_equal(A[num_vertices-1,:],A[num_vertices,:]):
        return A[num_vertices-1,:]
    else:
        #Return -1 if there is cycle
        return -1

```

```python
def read_input_johnson(filename):
    first = True
    with open('week13_file/'+filename) as f:
        G = {}
        G_rev = {}
        for line in f:
            line = line.split() # to deal with blank 
            if line and not first:            # lines (ie skip them)
                v = int(line[0])
                w = int(line[1])
                length = int(line[2])
                if G.get(v) == None:
                    G[v] = [(v,w,length)]
                else:
                    G[v].append((v,w,length))
                    
                    
                if G_rev.get(w) == None:
                    G_rev[w] = [(v,w,length)]
                else:
                    G_rev[w].append((v,w,length))
            elif line and first:
                num_vertices = int(line[0])
                num_edges = int(line[1])
                first = False
                
                    
    return G,G_rev,num_vertices,num_edges
```

```python
def bellman_ford(G_rev,num_vertices,num_edges,source):
    
    #num_vertices+1 to check cycle
    A = np.zeros((num_vertices+1,num_vertices))
    
    #Intialize base cases
    A[0,:] = np.inf
    A[0,source] = 0
    
    #iterate num_vertices+1 times to check cycle
    for i in range(1,num_vertices+1):
        for v in G_rev.keys():
            wv = []
            for edges in G_rev[v]:
                #print(v,edges)
                #print(edges)
                w = edges[0]
                cwv = edges[2]
                wv.append(A[i-1,w]+cwv)
            #print(wv)
            A[i,v] = min(A[i-1,v],min(wv))
    
    #print(A)
    if np.array_equal(A[num_vertices-1,:],A[num_vertices,:]):
        return A[num_vertices-1,:]
    else:
        #Return -1 if there is cycle
        return -1
```

```python
def dijkstra_algorithm_heap(G,num_vertices,source):
    '''Invariants: 1) elements in heap = vertices V-X. 2) for v not element X. Key[v] = smallest
    Djikstra Greedy score for an edge (u,v) in E with u in X (or +infinity it no such edges exist)'''
    
    X = [] #Vertices processed so far
    A = np.full((num_vertices+1),1000000)
    #A = [1000000]*(num_vertices+1)  #n size of Graph G
    A[source] = 0 #A is computed shortest path distances
    hd = heapdict()
    
    for i in range(num_vertices+1):
        hd[i] = 1000000
    hd[source] = 0
    
    while any(hd):
        w,distance = hd.popitem()
        X.append(w)
        A[w] = distance
        #maintaining second invariant, when w extracted from heap (i.e, added to X)
        #for each edge (w,v) in E
        if G.get(w) is not None:
            for (w,v,weight_wv) in G[w]: 
                #if v in V-X (i.e, in heap)
                if hd.get(v) != None:
                    #recompute key[v] = min{key[v],A[w]+Iwv} and reiinsert v into the heap
                    hd[v] = min(hd[v],A[w]+weight_wv)
    return A
```

```python
def johnson_algorithm(filename):
    G,G_rev,num_vertices,num_edges = read_input_johnson(filename)
    #index vertices 1 - num_vertices (inclusive)
    #form G' by adding a new vertex s and a new edge (s,v) with length 0 for each v element G
    s = 0
    
    G_rev[1] = [(s,1,0)]
    #index vertices 0 - num_vertices (inclusive)
    for i in range(1,num_vertices+1):
        if G_rev.get(i) == None:
            G_rev[i] = [(s,i,0)]
        else:
            G_rev[i].append((s,i,0))
    
    #Run Bellman-Ford on G' with source vertex S.
    
    bellman_result = bellman_ford(G_rev,num_vertices+1,num_edges+num_vertices,s)
    
    #If B-F detects a negative-cost cycle in G', halt and report this
    if type(bellman_result) == bool:
        return False
    
    p = {}
    
    #For each v element G, define pv = length of a shortest s -> v path in G'. 
    for v in range(len(bellman_result)):
        p[v] = bellman_result[v]

    
    G_reweight = {}
    
    for edges in G.values():
        for edge in edges:
            u = edge[0]
            v = edge[1]
            weight = edge[2]
            #For each edge e = (u,v) element G, define c'e = ce + pu-pv
            new_weight = weight + p[u] - p[v]
            if G_reweight.get(u) == None:
                G_reweight[u] = [(u,v,new_weight)]
            else:
                G_reweight[u].append((u,v,new_weight))

    shortest_path = np.zeros((num_vertices+1,num_vertices+1))

    #For each vertex u of G: Run Dijkstra's algorithm in G, with edge lengths {c'e}, with source vertex u
    #, to compute the shortest-path distance d'(u,v) for each v element G
    for source in range(1,num_vertices+1):
        A = dijkstra_algorithm_heap(G_reweight,num_vertices,source)
        
        #For each pair u,v element G, return the shortest-path distance d(u,v) := d'(u,v) -pu +pv
        for i in range(len(A)):
            A[i] = A[i] - p[source] + p[i]
        shortest_path[source] = A
        
    return np.min(shortest_path)
```
